```python
import pandas as pd
```

```python
stats = pd.read_csv('./data/padel_player_dataset.csv')
```

```python
intermediate_level = stats[stats['Skill Level'] == 'Intermediate']
intermediate_level
```

```python
intermediate_level_male = intermediate_level[intermediate_level['Gender'] == 'Male']
intermediate_level_male
```

```python
intermediate_level_male = intermediate_level_male.sort_values(by='Age Group', ascending = True)
intermediate_level_male
```

```python
intermediate_level_male.to_csv('intermediate_male.csv')
```

```python
intermediate_level_female = intermediate_level[intermediate_level['Gender'] == 'Female']
intermediate_level_female
```

```python
intermediate_level_female = intermediate_level[intermediate_level['Skill Level'] == 'Intermediate']
intermediate_level_female
```

```python
intermediate_level_female = intermediate_level_female.sort_values(by= 'Age Group', ascending = True)
intermediate_level_female
```

```python
intermediate_level_female.to_csv('intermediate_female.csv')
```
