```python
import pandas as pd

# Lista para almacenar los DataFrames de cada archivo CSV
dfs = []

# Iterar sobre todos los archivos CSV
for i in range(1, 21):
    # Construir la ruta al archivo CSV
    ruta_archivo = f'../data_padel/padel_player_dataset_{i:03}.csv'
    
    # Leer el archivo CSV y agregarlo al DataFrame
    df = pd.read_csv(ruta_archivo)
    dfs.append(df)

# Concatenar todos los DataFrames en uno solo
test = pd.concat(dfs, ignore_index=True)
```

```python
test.info()
```

```python
# Aquí asumimos que 'df' es el DataFrame resultante de la concatenación de los CSV

# Crear una columna 'Partido' con números consecutivos para cada grupo de ID
test['Partido'] = test.groupby('Player ID').cumcount() + 1
```

```python
import numpy as np
import pandas as pd

# Supongamos que tenemos 1000 IDs
total_ids = 1000

# Calculamos la cantidad de grupos de edad
cantidad_grupos_edad = 5  # Cambia esto según tus necesidades

# Dividimos los IDs en grupos de edad
ids_por_grupo = total_ids // cantidad_grupos_edad

# Creamos los rangos de edad
rangos_edad = ['<18', '18-30', '31-45', '46-60', '>60']

# Repetimos los rangos de edad según la cantidad de IDs por grupo
edades = np.repeat(rangos_edad, ids_por_grupo)

# Creamos un DataFrame con los IDs y las edades asignadas
df_edades = pd.DataFrame({'Player ID': range(1, total_ids + 1), 'Edad': edades})

# Visualizamos el resultado
print(df_edades)
```

```python
merged_df = pd.merge(test, df_edades, on='Player ID', how='inner')

# Mostrar el DataFrame fusionado
merged_df
```

```python
import pandas as pd
import numpy as np

# Definimos el número total de IDs
total_ids = 1000

# Creamos la lista de géneros (masculino y femenino)
generos = ['Masculino', 'Femenino']

# Calculamos cuántas veces debemos repetir cada género para obtener 20,000 registros
repeticiones_por_genero = total_ids * 20 // len(generos)

# Repetimos la lista de géneros para igualar la longitud de 20,000 registros
generos_repetidos = np.repeat(generos, repeticiones_por_genero)

# Ajustamos el número total de IDs para que coincida con la longitud de los géneros repetidos
total_ids = len(generos_repetidos)

# Creamos el DataFrame con los IDs y los géneros asignados
df_genero = pd.DataFrame({'Player ID': range(1, total_ids + 1),
                          'Género': generos_repetidos})

# Mostramos el DataFrame resultante
print(df_genero)
```

```python
merged_df = pd.merge(merged_df, df_genero, on='Player ID', how='left')

# Mostramos el resultado del merge
merged_df
```
