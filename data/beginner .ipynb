```python
stats = pd.read_csv('./data/padel_player_dataset.csv')
```

```python
stats.info()
```

```python
stats.head(10)
```

```python
beginner_level = stats[stats['Skill Level'] == 'Beginner']
beginner_level
```

```python
beginner_level = beginner_level.sort_values(by= 'Skill Level', ascending = True)
beginner_level
```

```python
beginner_level_male = beginner_level[beginner_level['Gender'] == 'Male']
beginner_level_male
```

```python
beginner_level_male = beginner_level_male.sort_values(by= 'Age Group', ascending = True)
beginner_level_male
```

```python
beginner_level_male.to_csv('beginner_male.csv')
```

```python
beginner_level_female = beginner_level[beginner_level['Gender'] == 'Female']
beginner_level_female
```

```python
beginner_level_female = beginner_level_female.sort_values(by= 'Age Group', ascending = True)
beginner_level_female
```

```python
beginner_level_female.to_csv('beginner_female.csv')
```
